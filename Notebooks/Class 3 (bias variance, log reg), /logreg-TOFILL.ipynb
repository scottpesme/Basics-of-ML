# Logistic Regression in 2D — Gradient Descent

This notebook creates two synthetic 2D datasets — one **linearly separable** and one **non‑separable** — and trains a logistic regression classifier on each **with vanilla gradient descent** on the **logistic loss**.

```python
import numpy as np
import matplotlib.pyplot as plt
```


## Data generators


```python
def make_linearly_separable(n_per_class=200):
    mean_pos = np.array([2.0, 2.0])
    mean_neg = np.array([-2.0, -2.0])
    cov = np.array([[0.6, 0.0],[0.0, 0.6]])
    X_pos = np.random.multivariate_normal(mean_pos, cov, size=n_per_class)
    X_neg = np.random.multivariate_normal(mean_neg, cov, size=n_per_class)
    X = np.vstack([X_pos, X_neg])
    y = np.hstack([np.ones(n_per_class), -np.ones(n_per_class)])
    return X, y

def make_nonseparable(n_per_class=200):
    mean_pos = np.array([0.8, 0.8])
    mean_neg = np.array([-0.8, -0.8])
    cov = np.array([[1.6, 0.4],[0.4, 1.6]])
    X_pos = np.random.multivariate_normal(mean_pos, cov, size=n_per_class)
    X_neg = np.random.multivariate_normal(mean_neg, cov, size=n_per_class)
    X = np.vstack([X_pos, X_neg])
    y = np.hstack([np.ones(n_per_class), -np.ones(n_per_class)])
    return X, y

def add_bias_column(Xs):
    # Append a column of ones as the last column (bias)
    ones = np.ones((Xs.shape[0], 1))
    return np.hstack([Xs, ones])
```

# Create datasets

```python
np.random.seed(2)

# Generate datasets
X_sep, y_sep = make_linearly_separable(n_per_class=200)
X_nonsep, y_nonsep = make_nonseparable(n_per_class=200)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot linearly separable
axes[0].scatter(X_sep[y_sep == 1, 0], X_sep[y_sep == 1, 1], label="+1")
axes[0].scatter(X_sep[y_sep == -1, 0], X_sep[y_sep == -1, 1], label="-1")
axes[0].set_title("Linearly separable data")
axes[0].set_xlabel("x₁")
axes[0].set_ylabel("x₂")
axes[0].legend()

# Plot non-separable
axes[1].scatter(X_nonsep[y_nonsep == 1, 0], X_nonsep[y_nonsep == 1, 1], label="+1")
axes[1].scatter(X_nonsep[y_nonsep == -1, 0], X_nonsep[y_nonsep == -1, 1], label="-1")
axes[1].set_title("Non-separable data")
axes[1].set_xlabel("x₁")
axes[1].set_ylabel("x₂")
axes[1].legend()

plt.tight_layout()
plt.show()
```


## Gradient descent optimiser


```python
def train_log_gd(Xtilde, y, lr=0.5, steps=1500):
    w = np.zeros(Xtilde.shape[1])
    ws, losses = [], []
    for t in range(steps):
        loss, gw = logistic_loss_and_grad(w, Xtilde, y)
        losses.append(loss)
        ws.append(w.copy())
        w -= lr * gw
    return np.array(ws), np.array(losses)
```


## Plot helpers


```python
def plot_training_loss(losses, title="Loss vs iterations", loglog=True):
    iters = np.arange(1, len(losses) + 1)
    plt.figure(figsize=(5, 4))
    if loglog:
        plt.loglog(iters, losses)
    else:
        plt.plot(iters, losses)
    plt.xlabel("Iteration")
    plt.ylabel("Training logistic loss")
    plt.title(title)
    plt.tight_layout()
    plt.show()


def plot_data_with_separator(Xs, y, w, title):
    # Xs is the 2D (standardised) feature matrix without the bias column
    plt.figure()
    X_pos = Xs[y == 1]
    X_neg = Xs[y == -1]
    plt.scatter(X_pos[:,0], X_pos[:,1], label="+1")
    plt.scatter(X_neg[:,0], X_neg[:,1], label="-1")
    # Decision boundary: w0*x + w1*y + w2*1 = 0  => y = -(w0*x + w2)/w1
    x_min, x_max = Xs[:,0].min() - 0.5, Xs[:,0].max() + 0.5
    xs = np.linspace(x_min, x_max, 200)
    w0, w1, wb = w[0], w[1], w[2]
    if abs(w1) > 1e-12:
        ys = -(w0 * xs + wb) / w1
        plt.plot(xs, ys, label="Separating hyperplane")
    else:
        if abs(w0) > 1e-12:
            xv = -wb / w0
            plt.axvline(x=xv, label="Separating hyperplane")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_loss_surface(w_traj, Xtilde, y, n_levels=20, title="Logistic loss surface and trajectory"):
    w_traj = np.asarray(w_traj)
    w_final = w_traj[-1]
    wb = w_final[2]  # hold bias fixed

    # Grid for (w0, w1)
    w0s = np.linspace(w_traj[:, 0].min() - 1, w_traj[:, 0].max() + 1, 120)
    w1s = np.linspace(w_traj[:, 1].min() - 1, w_traj[:, 1].max() + 1, 120)
    W0, W1 = np.meshgrid(w0s, w1s)

    def loss_at(w0, w1, wb, Xtilde, y):
        w = np.array([w0, w1, wb])
        z = Xtilde @ w
        yz = y * z
        data = np.logaddexp(0.0, -yz).mean()
        return data

    # Compute loss grid
    Z = np.zeros_like(W0)
    for i in range(W0.shape[0]):
        for j in range(W0.shape[1]):
            Z[i, j] = loss_at(W0[i, j], W1[i, j], wb, Xtilde, y)

    # Plot
    fig, ax = plt.subplots(figsize=(5, 4))
    cs = ax.contour(W0, W1, Z, levels=n_levels)
    ax.clabel(cs, inline=True, fontsize=8)
    ax.plot(w_traj[:, 0], w_traj[:, 1], '-o', ms=2, lw=1, label='trajectory')
    ax.scatter(w_traj[0, 0], w_traj[0, 1], color='red', label='start', zorder=3)
    ax.scatter(w_traj[-1, 0], w_traj[-1, 1], color='green', label='end', zorder=3)
    ax.set_xlabel('w₀')
    ax.set_ylabel('w₁')
    ax.legend()
    ax.set_title(title)
    plt.tight_layout()
    plt.show()
```


## Experiment 1 — Linearly separable data


```python
X_sep_tilde = add_bias_column(X_sep)

# Run gradient descent and keep all iterates
ws_sep, losses_sep = train_log_gd(X_sep_tilde, y_sep, steps=1500)

# Extract final parameters
w_sep_final = ws_sep[-1]
```

```python
plot_training_loss(losses_sep, loglog = True)
```

```python
plot_loss_surface(ws_sep, X_sep_tilde, y_sep)
```

```python
plot_data_with_separator(X_sep, y_sep, w_sep_final, "Separable — data + learned separator")
```
